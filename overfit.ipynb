```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
data = pd.read_csv(r"C:\Users\polam\OneDrive\Desktop\New folder\cleaned_data.csv",sep=',')
```

```python
df = data.copy()
```

```python
x = df.drop("Car", axis=1)
y = df["Car"]
```

```python
y = pd.get_dummies(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
```

### scaling

```python
scalar = StandardScaler()
x_train = scalar.fit_transform(x_train)
x_test = scalar.transform(x_test)
```

### model building

```python
model = Sequential([
    Dense(256, activation='relu', input_shape=(5900,)),
    Dropout(0.5),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(8, activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

### model evaluation

```python
epochs_list = [10, 20, 40, 60, 70, 100]

for epochs in epochs_list:
    print(f"\nTraining for {epochs} epochs:")
    history = model.fit(x_train, y_train, epochs=epochs, batch_size=32, validation_data=(x_test, y_test), verbose=1)
    
    
    max_accuracy = max(history.history['accuracy']) 
    best_epoch = history.history['accuracy'].index(max_accuracy) + 1   

    print(f"Highest Accuracy for {epochs} epochs: {max_accuracy:.4f} at Epoch {best_epoch}")
```

### regularization

```python
from tensorflow.keras import regularizers
```

```python
model = Sequential([
    Dense(256, activation='relu', input_shape=(5900,), kernel_regularizer=regularizers.l2(0.01)),
    Dropout(0.5),
    Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.01)),
    Dropout(0.5),
    Dense(8, activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
epochs_list = [10, 20, 40, 60, 70, 100]

for epochs in epochs_list:
    print(f"\nTraining for {epochs} epochs:")
    history = model.fit(x_train, y_train, epochs=epochs, batch_size=32, validation_data=(x_test, y_test), verbose=1)
    
    # Find the epoch with the highest accuracy for this training
    max_accuracy = max(history.history['accuracy'])
    best_epoch = history.history['accuracy'].index(max_accuracy) + 1  # 1-based index

    print(f"Highest Accuracy for {epochs} epochs: {max_accuracy:.4f} at Epoch {best_epoch}")
```
